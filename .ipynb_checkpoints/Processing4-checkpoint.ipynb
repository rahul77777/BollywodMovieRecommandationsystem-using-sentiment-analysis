```python
import pandas as pd
import numpy as np
```

```python
link = "https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2016"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
df
```

```python
df_2016 = df[['Title','Director','Cast','Genre']]
```

```python
df_2016
```

```python
df_2016 = df_2016.rename(columns={'Title':'title'})
df_2016 = df_2016.rename(columns={'Director':'directors'})
df_2016 = df_2016.rename(columns={'Cast':'actors'})
df_2016 = df_2016.rename(columns={'Genre':'genre'})
```

```python
df_2016
```

```python
df_2016['directors'] = df_2016['directors'].replace(np.nan, 'unknown')
df_2016['actors'] = df_2016['actors'].replace(np.nan, 'unknown')
df_2016['genre'] = df_2016['genre'].replace(np.nan, 'unknown')
```

```python
df_2016['title'] = df_2016['title'].str.lower()
```

```python
df_2016['comb'] = df_2016['directors'] + ' ' + df_2016['actors'] + ' '+ df_2016['genre']
```
