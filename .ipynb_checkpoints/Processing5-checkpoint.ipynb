```python
import pandas as pd
import numpy as np
```

```python
link = "https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2017"
df1 = pd.read_html(link, header=0)[2]
df2 = pd.read_html(link, header=0)[3]
df3 = pd.read_html(link, header=0)[4]
df4 = pd.read_html(link, header=0)[5]
```

```python
df = df1.append(df2.append(df3.append(df4,ignore_index=True),ignore_index=True),ignore_index=True)
```

```python
df
```

```python
df_2017 = df[['Title','Director','Cast','Genre']]
```

```python
df_2017
```

```python
df_2017 = df_2017.rename(columns={'Title':'title'})
df_2017 = df_2017.rename(columns={'Director':'directors'})
df_2017 = df_2017.rename(columns={'Cast':'actors'})
df_2017 = df_2017.rename(columns={'Genre':'genre'})
```

```python
df_2017
```

```python
df_2017['directors'] = df_2017['directors'].replace(np.nan, 'unknown')
df_2017['actors'] = df_2017['actors'].replace(np.nan, 'unknown')
df_2017['genre'] = df_2017['genre'].replace(np.nan, 'unknown')
```

```python
df_2017['title'] = df_2017['title'].str.lower()
```

```python
df_2017['comb'] = df_2017['directors'] + ' ' + df_2017['actors'] + ' '+ df_2017['genre']
```
