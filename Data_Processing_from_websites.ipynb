# Reading Movies data from Wikipedia for the year of 2015

```python
link_2015 = "https://en.wikipedia.org/wiki/List_of_Bollywood_films_of_2015"

dataframe1 = pd.read_html(link_2015, header=0)[2]
dataframe2 = pd.read_html(link_2015, header=0)[3]
dataframe3 = pd.read_html(link_2015, header=0)[4]
dataframe4 = pd.read_html(link_2015, header=0)[5]

dataframe = pd.concat([dataframe1,dataframe2,dataframe3,dataframe4], ignore_index = True, sort = False)
```

```python
dataframe
```

## Choosing only required columns

```python
dataframe_2015 = dataframe[['Title','Director','Cast','Genre']]
```

```python
dataframe_2015
```

### Renaming the columns

```python
dataframe_2015 = dataframe_2015.rename(columns={'Title':'title','Director':'directors','Cast':'actors','Genre':'genre'})
```

```python
dataframe_2015
```

## Replacing nan values with unknown

```python
dataframe_2015['directors'] = dataframe_2015['directors'].replace(np.nan, 'unknown')
dataframe_2015['actors'] = dataframe_2015['actors'].replace(np.nan, 'unknown')
dataframe_2015['genre'] = dataframe_2015['genre'].replace(np.nan, 'unknown')
```

```python
dataframe_2015['directors'] = dataframe_2015['directors'].str.replace(',', ' ')
dataframe_2015['actors'] = dataframe_2015['actors'].str.replace(',', ' ')
dataframe_2015['genre'] = dataframe_2015['genre'].str.replace(',', ' ')
```

```python
dataframe_2015['title'] = dataframe_2015['title'].str.lower()
```

```python
dataframe_2015['cast'] = dataframe_2015['directors'] + ' ' + dataframe_2015['actors']+' '+ dataframe_2015['genre']
```

```python
dataframe_2015
```

```python
dataframe_2015.to_csv('2015_movies.csv',index=False)
```
